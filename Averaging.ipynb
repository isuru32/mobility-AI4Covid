### avg1 = PP – [FP+SP+TP]/3

```python
for i in avgCols:
    avg1[i] = round((all[i+'_PP'] - (all[i+'_FP'] + all[i+'_SP'] + all[i+'_TP'])/3),3)
    avg2[i] = round((all[i+'_PP'] - ((all[i+'_FP'] + all[i+'_SP'])/2 + all[i+'_TP'])/2),3)
    avg3[i] = round((all[i+'_PP'] - ((all[i+'_SP'] + all[i+'_TP'])/2 + all[i+'_FP'])/2),3)
```
