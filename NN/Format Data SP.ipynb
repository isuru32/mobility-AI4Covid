```python
f_choice = ['NA']*1898
s_choice = ['NA']*1898
t_choice = ['NA']*1898

print(f_choice)
```

```python
sof_dict = sof1.to_dict(orient='index')
```

```python
sof_dict
```

```python
# Sort the inner dictionaries
sof_dict = {key: dict(sorted(value.items(), key=lambda item: item[1], reverse=True))
                       for key, value in sof_dict.items()}

print(sof_dict)
```

```python
# sof_dict_df = pd.DataFrame.from_dict(sof_dict, orient='index')
# sof_dict_df
```

```python
# num = 0
# for outer_key, inner_dict in sof_dict.items():
#     # print(f"Outer key: {outer_key}")
#     d = list(inner_dict)
#     f_choice[num] = d[0]
#     s_choice[num] = d[1]
#     t_choice[num] = d[2]
#     num += 1
# print(f_choice)
```

```python
num = 0
for outer_key, inner_dict in sof_dict.items():
    # print(f"Outer key: {outer_key}")
    first_inner_key = next(iter(inner_dict))
    if inner_dict[first_inner_key] == 1:
        f_choice[num] = 'No'
        s_choice[num] = 'No'
        t_choice[num] = 'No'
        # print('1',num)
        # print(f_choice)
    else:
        d = list(inner_dict)
        f_choice[num] = d[0]
        s_choice[num] = d[1]
        t_choice[num] = d[2]
        # print('2',num)

    num += 1
print(f_choice)
```

```python
sof['f_choice'] = f_choice
sof['s_choice'] = s_choice
sof['t_choice'] = t_choice
```
