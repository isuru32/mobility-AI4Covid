```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
avgCols = ['sof_neighborhood_shop', 'sof_main_super_market', 'sof_mini_super_market', 'sof_home_delivery_or_mobile_shops', 'sof_own_field', 'sof_govt_ngo', 'sof_online', 'sof_neighbors', 'sof_fairs', 'fin_bank_city', 'fin_bank_near', 'fin_local_financial', 'fin_atm_city', 'fin_atm_near', 'fin_online_banking', 'fin_post_office', 'lei_trips', 'lei_recreational', 'lei_social_gathering', 'mot_public', 'mot_shared_vehi', 'mot_hired', 'mot_own_vehicle', 'mot_bicycle', 'mot_walking', 'w_work_from_home', 'w_come_from_home', 'w_come_from_renthouse', 'w_come_from_commonhostel', 'wmot_public', 'wmot_shared', 'wmot_hired', 'wmot_own_vehicle', 'wmot_bicycle', 'wmot_walking', 'w_alone_home','w_with_few_known','w_with_few_known_lim_unknown','w_with_few_irre_unknown','w_with_large_irre_unknown']
wave = ['PP','FP','SP','TP']
```

```python
rows = 2
cols = 2
fig, axes = plt.subplots(rows,cols, figsize=(5,5))
x = -1
plt.suptitle('sof_neighborhood_shop')
for i in range(rows):
    for j in range(cols):
        x = x + 1
        # print('sof_neighborhood_shop'+'_'+wave[x])
        # print(i,j)
        axes[i][j].pie(all['sof_neighborhood_shop'+'_'+wave[x]].value_counts(dropna=False).values,labels=all['sof_neighborhood_shop'+'_'+wave[x]].value_counts().index, autopct='%1.1f%%', startangle = 90, wedgeprops = {"edgecolor" : "black", 
                      'linewidth': 1,  
                      'antialiased': True}) 
  
        # Equal aspect ratio ensures 
        # that pie is drawn as a circle. 
        plt.axis('equal')
        axes[i][j].set_title(wave[x])
plt.show()
```

```python
rows = 2
cols = 2
fig, axes = plt.subplots(rows,cols, figsize=(5,5))
x = -1
plt.suptitle('sof_govt_ngo')
colors={'Never':'cornflowerblue', 
        'Less Often':'gold',
        'Often': 'green',
        'Very Often': 'red'}
for i in range(rows):
    for j in range(cols):
        x = x + 1
        # print('sof_neighborhood_shop'+'_'+wave[x])
        # print(i,j)
        axes[i][j].pie(all['sof_govt_ngo'+'_'+wave[x]].value_counts(dropna=False).values,labels=all['sof_govt_ngo'+'_'+wave[x]].value_counts().index, autopct='%1.1f%%', startangle = 90, wedgeprops = {"edgecolor" : "black", 
                      'linewidth': 1,  
                      'antialiased': True},
                      colors=[colors[v] for v in all['sof_govt_ngo'+'_'+wave[x]].value_counts().keys()]) 
  
        # Equal aspect ratio ensures 
        # that pie is drawn as a circle. 
        plt.axis('equal')
        axes[i][j].set_title(wave[x])
plt.show()
```
